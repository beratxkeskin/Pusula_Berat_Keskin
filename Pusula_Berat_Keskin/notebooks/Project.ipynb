```python
df["Cinsiyet"] = df.groupby("HastaNo")["Cinsiyet"].transform(lambda x: x.ffill().bfill())
```

```python
df['Cinsiyet'].isnull().sum()
```

```python
from sklearn.impute import KNNImputer
```

```python
df["Cinsiyet_num"] = df["Cinsiyet"].map({"Kadın":0, "Erkek":1})

knn = KNNImputer(n_neighbors=5)
df[["Cinsiyet_num"]] = knn.fit_transform(df[["Cinsiyet_num"]])

df["Cinsiyet"] = np.where(df["Cinsiyet_num"] < 0.5, "Kadın", "Erkek")
df = df.drop(columns=["Cinsiyet_num"])
```

```python
df['Cinsiyet'].isnull().sum()
```

```python
df['KanGrubu'].isnull().sum()
```

```python
df["KanGrubu"] = df.groupby("HastaNo")["KanGrubu"].transform(lambda x: x.ffill().bfill())
```

```python
df['KanGrubu'].isnull().sum()
```

```python
df["KanGrubu"] = df["KanGrubu"].fillna("Unknown")
```

```python
df['KanGrubu'].isnull().sum()
```

```python
df["KronikSayisi"] = df["KronikHastalik"].apply(
    lambda x: len(str(x).split(",")) if pd.notna(x) else 0
)
```

```python
top_10_diseases = [
    "Aritmi", "Hiportiroidizm", "Limb-Girdle Musküler Distrofi",
    "Astım", "Hipertiroidizm", "Myastenia gravis",
    "Diyabet", "Duchenne Musküler Distrofisi",
    "Fascioscapulohumeral Distrofi", "Kalp yetmezliği"
]

for disease in top_10_diseases:
    df[disease] = df["KronikHastalik"].apply(
        lambda x: 1 if pd.notna(x) and disease in str(x) else 0
    )
```

```python
df[["KronikHastalik", "KronikSayisi"] + top_10_diseases].head()
```

```python
df["Bolum"].value_counts(dropna=False)
```

```python
df["Bolum"] = df["Bolum"].fillna("Unknown")
```

```python
df["Alerji"].value_counts(dropna=False)
```

```python
df["Alerji"] = df["Alerji"].str.strip()  
df["Alerji"] = df["Alerji"].str.title()  

replace_dict = {
    "Polen": "Polen",
    "Toz": "Toz",
    "Sucuk": "Sucuk",
    "Novalgin": "Novalgin",
    "Volteren": "Voltaren",   
    "Voltaren": "Voltaren",
    "Gripin": "Gripin",
    "Grıpın": "Gripin",       
    "Arveles": "Arveles",
    "Coraspin": "Coraspin"
}

df["Alerji"] = df["Alerji"].replace(replace_dict, regex=True)
```

```python
df["Alerji"] = df["Alerji"].fillna("Yok")

all_allergies = []
for entry in df["Alerji"]:
    parts = [x.strip().title() for x in str(entry).split(",")]  
    all_allergies.extend(parts)

allergy_counts = pd.Series(all_allergies).value_counts()

top_10_allergies = allergy_counts.head(10)
top_10_allergies
```

```python
df["AlerjiSayisi"] = df["Alerji"].apply(
    lambda x: 0 if x == "Yok" else len(str(x).split(","))
)

important_allergies = ["Polen", "Toz", "Novalgin", "Sucuk", 
                       "Voltaren", "Arveles", "Coraspin", 
                       "Yer Fıstığı", "Gripin"]

for allergy in important_allergies:
    df[allergy] = df["Alerji"].apply(
        lambda x: 1 if allergy in str(x) else 0
    )
```

```python
df['Tanilar'].value_counts()
```

```python
def clean_tanilar(x):
    if pd.isna(x):
        return ["Bilinmiyor"]
    diagnoses = [d.strip().title() for d in str(x).replace(",,", ",").split(",") if d.strip()]
    return diagnoses if diagnoses else ["Bilinmiyor"]

df["Tanilar_list"] = df["Tanilar"].apply(clean_tanilar)
```

```python
df["TaniSayisi"] = df["Tanilar_list"].apply(len)
```

```python
all_diagnoses = pd.Series([d for sublist in df["Tanilar_list"] for d in sublist])
top_diagnoses = all_diagnoses.value_counts().head(15).index
```

```python
top_diagnoses
```

```python
for diag in top_diagnoses:
    df[f"Tani_{diag.replace(' ', '_')}"] = df["Tanilar_list"].apply(lambda x: 1 if diag in x else 0)
```

```python
df
```

```python
df["TedaviAdi"].value_counts().head(20)
```

```python
def clean_tedavi(x):
    if pd.isna(x):
        return ["Bilinmiyor"]
    tedaviler = [t.strip().title() for t in str(x).replace("--", "-").split("-") if t.strip()]
    return tedaviler if tedaviler else ["Bilinmiyor"]
```

```python
df["TedaviAdi_list"] = df["TedaviAdi"].apply(clean_tedavi)
```

```python
df["TedaviSayisi"] = df["TedaviAdi_list"].apply(len)
```

```python
df["TedaviSayisi"].value_counts()
```

```python
all_treatments = pd.Series([t for sublist in df["TedaviAdi_list"] for t in sublist])
top_treatments = all_treatments.value_counts().head(20)

top_treatments
```

```python
top_treatments = top_treatments.drop(["1", "2"], errors="ignore")
top_treatments
```

```python
threshold = 50

tedavi_counts = df["TedaviAdi"].value_counts()

common_tedaviler = tedavi_counts[tedavi_counts >= threshold].index

for tedavi in common_tedaviler:
    col_name = "Tedavi_" + tedavi.replace(" ", "").replace("+", "").replace("-", "").replace("İ", "I").replace("ı","i")
    df[col_name] = df["TedaviAdi"].apply(lambda x: 1 if tedavi in str(x) else 0)

df["Tedavi_Other"] = df["TedaviAdi"].apply(lambda x: 1 if x not in common_tedaviler else 0)
```

```python
df["TedaviSuresi"] = df["TedaviSuresi"].str.replace(" Seans", "", regex=False).astype(int)

print(df["TedaviSuresi"].describe())
```

```python
df["UygulamaYerleri"] = df["UygulamaYerleri"].str.replace(" Bölgesi", "", regex=False)

print(df["UygulamaYerleri"].value_counts().head(20))
```

```python
df["UygulamaYerleri"] = df["UygulamaYerleri"].str.split(",")

df["UygulamaYerleri"].head(10)
```

```python
uygulama_counts = df["UygulamaYerleri"].explode().value_counts()
top_uygulamalar = uygulama_counts[uygulama_counts >= 50].index.tolist()

for yer in top_uygulamalar:
    df[f"Uyg_{yer.replace(' ', '_')}"] = df["UygulamaYerleri"].apply(
        lambda x: 1 if isinstance(x, list) and yer in x else 0
    )
```
